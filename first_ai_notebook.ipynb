```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Student": ["Avishkar", "Om", "Aayush", "Sujal"],
    "Marks": [85, 78, 62, 74]
})

print("Data Table:")
print(data)

data.plot(kind="bar", x="Student", y="Marks", title="Student Marks Comparison")
plt.show()
```
